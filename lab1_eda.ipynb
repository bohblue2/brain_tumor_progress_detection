```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import glob
import pydicom
from pydicom.data import get_testdata_files
import os

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Tạo figure để hiển thị hình ảnh
fig, axes = plt.subplots(4, 6, figsize=(15, 10))

# Duyệt qua num từ 01 đến 24
for i, num in enumerate(range(1, 25)):
    # Định dạng số thành hai chữ số (01, 02, ...)
    file_num = f"{num:02}"
    try:
        # Đọc file DICOM
        ds = pydicom.dcmread(f'datasets/Brain-Tumor-Progression/PGBM-001/04-02-1992-FH-HEADBrain Protocols-79896/11.000000-T1post-80644/1-{file_num}.dcm')
        # Hiển thị ảnh trên subplot
        ax = axes[i // 6, i % 6]
        ax.imshow(ds.pixel_array, cmap=plt.cm.gray)
        ax.set_title(f"Image {file_num}")
        ax.axis('off')  # Tắt trục
    except Exception as e:
        print(f"Error loading file 1-{file_num}.dcm: {e}")

# Điều chỉnh khoảng cách giữa các hình
plt.tight_layout()
plt.show()
```
